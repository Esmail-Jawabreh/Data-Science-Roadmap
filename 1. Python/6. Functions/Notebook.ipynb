### simple function

```python
def greet(name):
    """This function greets the person passed as parameter"""
    print(f"Hello, {name}!")

greet("Esmail")
```

### Function with return value

```python
def add(a, b):
    return a + b

result = add(5, 3)
print("Sum:", result)
```

### Function with default argument

```python
def power(base, exponent=2):
    return base ** exponent

print(power(3))
print(power(3, 3))
```

### Keyword arguments

```python
def describe_person(name, age):
    print(f"{name} is {age} years old.")

describe_person(age=25, name="Esmail")
```

### Lambda function

```python
square = lambda x: x ** 2
print("Square of 5:", square(5))
```

### Variable scope example

```python
x = 10

def modify():
    x = 5  # local variable
    print("Inside function:", x)

modify()
print("Outside function:", x)
```

### Factorial

```python
def factorial(n):
    """Calculate factorial of n (n!)."""
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

### Palindrome

```python
def is_palindrome(s):
    """Check if string s is a palindrome."""
    return s == s[::-1]

print(is_palindrome("radar"))
print(is_palindrome("python"))
```

### global & nonlocal

```python
def sum_all(*args):
    """Sum all arguments."""
    return sum(args)

print(sum_all(1, 2, 3))       # 6
print(sum_all(5, 10, 15, 20)) # 50

def print_info(**kwargs):
    """Print key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Esmail", age=25, city="Amman")
```

### Docstrings

```python
def greet(name):
    """
    Greet a person by their name.

    Parameters:
    name (str): The name of the person to greet.

    Returns:
    None
    """
    print(f"Hello, {name}!")

greet("Esmail")
```

### Many arguments

```python
def sum_all(*args):
    """Sum all arguments."""
    return sum(args)

print(sum_all(1, 2, 3))
print(sum_all(5, 10, 15, 20))

def print_info(**kwargs):
    """Print key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Esmail", age=25, city="Amman")
```
