# Comprehensions & Lambda


- Concise and efficient ways to create lists and dictionaries in Python using comprehensions, plus anonymous functions (lambda) for short operations.


### List Comprehensions


```python
squares = [x**2 for x in range(10)]
print(squares)
```

```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)
```

### Dictionary Comprehensions

```python
square_dict = {x: x**2 for x in range(6)}
print(square_dict)
```

```python
even_square_dict = {x: x**2 for x in range(6) if x % 2 == 0}
print(even_square_dict)
```

# Lambda Functions

```python
add_ten = lambda x: x + 10
print(add_ten(5))
```

```python
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
print(squared)
```

```python
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```
