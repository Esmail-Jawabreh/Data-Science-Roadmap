# For Loop example

```python
print("For Loop Example:")
for i in range(5):
    print("Iteration:", i)
```

# While Loop example

```python
print("\nWhile Loop Example:")
count = 0
while count < 5:
    print("Count is:", count)
    count += 1
```

# Break statement example

```python
print("\nBreak Statement Example:")
for i in range(10):
    if i == 3:
        print("Breaking the loop at i =", i)
        break
    print(i)
```

# Continue statement example

```python
print("\nContinue Statement Example:")
for i in range(5):
    if i == 2:
        print("Skipping iteration at i =", i)
        continue
    print(i)
```

# Else clause with loops

- **Else with loops**: The `else` block runs after the loop completes normally (without encountering a `break`). It can be used with both `for` and `while` loops.


```python
for i in range(5):
    print(i)
else:
    print("Loop completed without break")
```

```python
count = 0
while count < 5:
    print(count)
    count += 1
else:
    print("While loop completed")
```

# Nested loops example: Multiplication Table

- **Nested loops**: A loop inside another loop, useful for iterating over multi-dimensional data or running repeated sequences within each iteration.


```python
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i} x {j} = {i*j}")
    print("-" * 10)
```

# Countdown with conditions

```python
for i in range(10, 0, -1):
    if i == 5:
        print("Halfway there!")
    print(i)
print("Blast off!")
```
