# Creating strings

```python
name = "Esmail"
greeting = 'Hello, ' + name + "!"
print(greeting)
```

# String methods


```python
print(name.upper())
print(name.lower())
print("  Hello World  ".strip())
```

# f-string formatting

```python
name = "Esmail"
age = 25
print(f"My name is {name} and I am {age} years old.")
```

# String indexing and slicing

```python
s = "Data Science"
print(s[0])      # First character
print(s[-1])     # Last character
print(s[1:5])    # Slice from index 1 to 4
```

# Escape characters and multiline strings

```python
multi_line = """This is
a multiline
string"""
print(multi_line)
```

```python
escaped = "This is a tab:\tand this is a newline:\nEnd"
print(escaped)
```
